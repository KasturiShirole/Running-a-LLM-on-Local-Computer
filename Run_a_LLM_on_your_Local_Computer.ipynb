```python
#  Load model and tokenizer with caching disabled
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_id = "microsoft/Phi-3-mini-4k-instruct"

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_id)

# Load model on GPU with appropriate settings
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    device_map="auto",  # will automatically use CUDA if available
    trust_remote_code=True
)

#  Disable caching to avoid DynamicCache error
model.config.use_cache = False
```

```python
# Prepare your prompt and inputs
prompt = "Create a funny joke about chickens."

# Tokenize input
inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
```

```python
#  Generate output safely
with torch.no_grad():
    output = model.generate(
        **inputs,
        max_new_tokens=100,
        do_sample=True,
        temperature=0.7,
        top_p=0.9
    )

# Decode the output
response = tokenizer.decode(output[0], skip_special_tokens=True)
print(response)
```
